```python
import seaborn as sns
```

```python
#Basic plotting in seaborn 
kt=sns.load_dataset('tips')     # when u used this there is lots of dataset available spo we pass tips dataset and printwe got the data set 
kt
    
```

```python
# create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip', data=kt)
plt.title("Scatter plot of Total bill vs Tip")
plt.show()
```

```python
#LINE PLOT
sns.lineplot(x='size', y='total_bill', data=kt)
plt.title("Line plot of total bill by size")
plt.show
```

```python
#categorical plot (like day, sex, time etc are categorical variable )
# some famous categorical plot is bar plot 
#Bar plot
sns.barplot(x='day', y='total_bill', data=kt)
plt.title('Bar plot of total bill By Day')
plt.show()
```

```python
#Box plot
sns.boxplot(x='day', y='total_bill', data=kt)
plt.title('Bax plot of total bill By Day')
plt.show()

```

```python
#violin plot
sns.violinplot(x='day', y ='total_bill', data=kt)
plt.title("Violine plot between day and total_bill")
plt.show()
```

```python
##Histrogram  
sns.histplot(kt['total_bill'],bins=10, kde=True)  # u sahould used false too for no histogram
```

```python
#kde plot
sns.kdeplot(kt['total_bill'], fill=True)
```

```python
#pairplot
sns.pairplot(kt)
```

```python
##heatmap
corr =kt[['total_bill','tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
